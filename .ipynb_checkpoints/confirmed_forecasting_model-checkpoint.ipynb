```python
# manipulte data before broadcasting 
confirmed_cases.columns = ['ds','y']
confirmed_cases
```

```python
# # Training confirmed_cases model
model = Prophet(interval_width=0.97)
# changepoints=['2009-03-09', '2010-07-02', '2011-09-26', '2012-03-20', '2010-04-06']
# n_changepoints=30
# ,daily_seasonality=True
model.fit(confirmed_cases)
```

```python
# prediction
future = model.make_future_dataframe(periods=7,freq='M')
forecast = model.predict(future)
forecast.tail()
```

```python
#to get the error
metric_df = forecast.set_index('ds')[['yhat']].join(confirmed_cases.set_index('ds').y).reset_index()
metric_df
```

```python
# nane as the future data first 
# was not exsist 
metric_df.dropna(inplace=True)
metric_df
```

```python
# mean_squared_error(metric_df.y, metric_df.yhat)
r2_score(metric_df.y, metric_df.yhat)
```

```python
model.plot(forecast)
for changepoint in model.changepoints:
    plt.axvline(changepoint,ls='--', lw=1)
plt.show()
```

```python
#the magnitude of changepoints
deltas = model.params['delta'].mean(0)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111)
ax.bar(range(len(deltas)), deltas)
ax.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2)
ax.set_ylabel('Rate change')
ax.set_xlabel('Potential changepoint')
fig.tight_layout()
```
